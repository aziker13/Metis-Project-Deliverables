```python
#going to next page test, doesn't need to run in final
#var=2
def get_reviews(url):
    chromedriver = "/home/aziker/WebDriver/chromedriver"
    driver = webdriver.Chrome(chromedriver)
    driver.get(url)
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    
    name = soup.find('h1', class_="css-1x9iesk").text
    
    reviews = [review.text.replace('\xa0', '') for review in soup.find_all('p', class_="comment__09f24__gu0rG css-1sufhje")]
    
    test = [i.text for i in soup.find_all('div', class_="border-color--default__09f24__NPAKY text-align--center__09f24__fYBGO")\
            if i.find('span', class_ = "css-1e4fdj9") != None]
    clicks =  int(test[0].split()[2])

    for _ in range(clicks-1):
        time.sleep(random.choice(range(6,10)))
        next_table = driver.find_element_by_xpath(f'//a[contains(@class, "next-link navigation-button__09f24__m9qRz css-1pxws0l")]')
        next_table.click()
        time.sleep(random.choice(range(3,7)))
        soup = BeautifulSoup(driver.page_source, 'html.parser')
        reviews.extend([review.text.replace('\xa0', '') for review in soup.find_all('p', class_="comment__09f24__gu0rG css-1sufhje")])
        
    names = [name] * len(reviews)
    data = {'shop name': names, 'reviews': reviews}
    df_reviews = pd.DataFrame(data = data)
    df_reviews.to_csv(f'{name}_reviews.csv')
    
    driver.quit()
    
    return 'download complete'
```

```python
get_reviews('https://www.yelp.com/biz/espresso-vivace-roasteria-seattle?osq=Coffee+Shop')
#it works!
```
